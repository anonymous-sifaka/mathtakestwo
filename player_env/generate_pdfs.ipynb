```python
import sys
import os

# Get path to the parent directory (i.e., project/)
project_root = os.path.abspath(os.path.join(os.getcwd(), '..'))
sys.path.append(project_root)

import matplotlib.pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
import mtt.environment as env
import numpy as np

 # A4 size in inches
a4_width, a4_height = 8.27*0.86, 11.69*0.86
    
for player_number in range(10):

    v_generator = env.QuizGenerator(player_number,'visualizer')
    d_generator = env.QuizGenerator(player_number,'describer')

    with PdfPages('./pdfs/player_' + str(player_number) + '_visualizer.pdf') as pdf:
        
        for page_start in range(0, 10, 5):  # Two pages if 10 questions
            fig, axes = plt.subplots(
                5, 4, figsize=(a4_width, a4_height),
                gridspec_kw={'wspace': 0.3, 'hspace': 0.6}  # spacing between subplots
            )

            for row in range(5):
                q_num = page_start + row + 1
                imgs = v_generator.get_next_practice(return_figure=True)

                for col in range(4):
                    ax = axes[row, col]
                    ax.imshow(imgs[col], cmap='gray')
                    ax.set_title(f"Prac Q {q_num} - Answer {col+1}", fontsize=8)
                    ax.axis('off')

            # Add margins via tight_layout
            fig.tight_layout(pad=1.0)
            pdf.savefig(fig)
            plt.close(fig)
        
        for page_start in range(0, 10, 5):  # Two pages if 10 questions
            fig, axes = plt.subplots(
                5, 4, figsize=(a4_width, a4_height),
                gridspec_kw={'wspace': 0.3, 'hspace': 0.6}  # spacing between subplots
            )

            for row in range(5):
                q_num = page_start + row + 1
                imgs = v_generator.get_next_test(return_figure=True)

                for col in range(4):
                    ax = axes[row, col]
                    ax.imshow(imgs[col], cmap='gray')
                    ax.set_title(f"Test Q {q_num} - Answer {col+1}", fontsize=8)
                    ax.axis('off')

            # Add margins via tight_layout
            fig.tight_layout(pad=1.0)
            pdf.savefig(fig)
            plt.close(fig)
        
    with PdfPages('./pdfs/player_' + str(player_number) + '_describer.pdf') as pdf:

        fig, axes = plt.subplots(
        5, 2, figsize=(a4_width, a4_height),
        gridspec_kw={'wspace': 0.3, 'hspace': 0.6}
        )

        q_num = 1
        for row in range(5):
            for col in range(2):
                img, answer = d_generator.get_next_practice(return_figure=True, return_answer=True)  # Single grayscale image
                ax = axes[row, col]
                ax.imshow(img, cmap='gray')
                ax.set_title(f"Prac Q:  {q_num} | Answer is {answer} ", fontsize=9)
                ax.axis('off')
                q_num += 1

        fig.tight_layout(pad=1.0)
        pdf.savefig(fig)
        plt.close(fig)
        
        fig, axes = plt.subplots(
        5, 2, figsize=(a4_width, a4_height),
        gridspec_kw={'wspace': 0.3, 'hspace': 0.6}
        )

        q_num = 1
        for row in range(5):
            for col in range(2):
                img, answer = d_generator.get_next_test(return_figure=True, return_answer=True)  # Single grayscale image
                ax = axes[row, col]
                ax.imshow(img, cmap='gray')
                ax.set_title(f"Test Question {q_num} | Answer is {answer}", fontsize=9)
                ax.axis('off')
                q_num += 1

        fig.tight_layout(pad=1.0)
        pdf.savefig(fig)
        plt.close(fig)

        
print('Finished')
```

```python
# A4 size in inches
a4_width, a4_height = 8.27*0.86, 11.69*0.86
example_generator = env.ExampleGenerator()

img_set = []
question_set = []
answer_set = []

for i in range(10):
    img, questions, answer = example_generator.show_example(return_figure=True, return_answer=True)
    answer = np.where(answer)[0][0] + 1
    img_set.append(img)
    question_set.append(questions)
    answer_set.append(answer)

for zz in range(2):
    with PdfPages('./pdfs/examples_'+str(zz)+'.pdf') as pdf:

        counter = 0
        for page_start in range(0, 10, 5):  # Two pages if 10 questions
            fig, axes = plt.subplots(
                5, 4, figsize=(a4_width, a4_height),
                gridspec_kw={'wspace': 0.3, 'hspace': 0.6}  # spacing between subplots
            )
            for row in range(5):

                q_num = page_start + row + 1
                imgs = question_set[counter]
                counter+=1

                for col in range(4):
                    ax = axes[row, col]
                    ax.imshow(imgs[col], cmap='gray')
                    ax.set_title(f"Example Q {q_num} - Answer {col+1}", fontsize=8)
                    ax.axis('off')

            # Add margins via tight_layout
            fig.tight_layout(pad=1.0)
            pdf.savefig(fig)
            plt.close(fig)

        counter = 0
        fig, axes = plt.subplots(
        5, 2, figsize=(a4_width, a4_height),
        gridspec_kw={'wspace': 0.3, 'hspace': 0.6}
        )
        q_num = 1
        for row in range(5):
            for col in range(2):
                img = img_set[counter]
                ax = axes[row, col]
                ax.imshow(img, cmap='gray')
                ax.set_title(f"Example Q {q_num} | answer is {answer_set[counter]}", fontsize=9)
                ax.axis('off')
                q_num += 1
                counter += 1 

        fig.tight_layout(pad=1.0)
        pdf.savefig(fig)
        plt.close(fig)
    
edge_set = example_generator.show_edge_cases(return_figure=True)

with PdfPages('./pdfs/edge_cases.pdf') as pdf:

    counter = 0
    fig, axes = plt.subplots(
    4, 1, figsize=(a4_width, a4_height),
    gridspec_kw={'wspace': 0.3, 'hspace': 0.6}
    )

    q_num = 1
    for row in range(4):
        img = edge_set[counter]
        counter += 1 
        ax = axes[row]
        ax.imshow(img, cmap='gray')
        ax.set_title(f"Edge Question {q_num}", fontsize=9)
        ax.axis('off')
        q_num += 1

    fig.tight_layout(pad=1.0)
    pdf.savefig(fig)
    plt.close(fig)
    
print('Finished')
```

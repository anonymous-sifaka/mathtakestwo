```python
#### Task Set Up ####

# Your task is to develop a language to communicate specific images to each other
# In this 1st learning phase you will work together to create a language
# In the next phase you will be seperated and only allowed to communicate via your language

#### Objective ####

# One person will be the Describer and be presented with an image
# This person must write a single communication to the Visualizer using your new language

# The Visualizer will be shown 4 images and asked - which one matches what the Desciber saw?
# The Visualizer must use the recieved communication to answer the question correctly

#### Language Rules ####

# Your language may only contain the following 8 characters:
# A, B, C, 0, 1, 2, +, *,

# Communications may only be 8 character long i.e. "AAAABBBB"

# Only one communication is allowed per image not two, not three, and certainly not four.

# If you break the rules in the practice or test phases you will be DISQUALIFIED!!!
# Good luck and may the TYPO'S favour you...
```
